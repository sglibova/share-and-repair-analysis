Note that Apple changes their response headers often enough that this script may not be able to authorize against their API by the time you run this notebook. Pre-collected review data can be found at `initial_needfinding/data/reviews`.

```python
def fetch_all_reviews(app_name: str, app_id: str, offset: str, country='us'):
    all_reviews = []
    user_agents = [
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 13_4) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/16.4 Safari/605.1.15',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36',
        ]
    token = get_token(country=country, app_name=app_name, app_id=app_id, user_agents=user_agents)
    
    while offset is not None:
        batch, offset, status = fetch_reviews(country=country, app_name=app_name, app_id=app_id, user_agents=user_agents, token=token, offset=offset)
        all_reviews.extend(batch)
        print(f"Total so far: {len(all_reviews)}")

    print(f"Done. Total reviews: {len(all_reviews)}")
    return all_reviews
```

```python
finditfixit = fetch_all_reviews(app_name='Find-It-Fix-It', app_id='568509551', offset='1')
```

```python
def write_review_csv(reviews: list, filename: str):
    if [review['attributes'] for review in reviews]:
        all_reviews = pd.DataFrame(reviews)
        reviews_df = pd.json_normalize(all_reviews['attributes'])
        print(f"Writing {filename}.csv...")
        reviews_df.to_csv(f"{filename}.csv", index=False)
        return reviews_df
    else:
        print(f"Reviews don't contain an 'attributes' property, returning as is...")
        return reviews
    
findit_df = write_review_csv(finditfixit, "../../initial_needfinding/data/reviews/apple/FindItFixIt")
seeclick_df = write_review_csv(seeclickfix, "../../initial_needfinding/data/reviews/apple/SeeClickFix")
sanjose_df = write_review_csv(sanjose311, "../../initial_needfinding/data/reviews/apple/SanJose311")
```
