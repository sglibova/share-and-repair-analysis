```python
from google_play_scraper import app, Sort, reviews_all
import pandas as pd
```

```python
finditfixit = 'gov.seattle.searequests'
sanjose311 = 'com.astcorporation.three11'
seeclickfix = 'com.seeclickfix.ma.android'
```

```python
finditfixit_results = reviews_all(
    finditfixit,
    sleep_milliseconds=10, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
sj311_results = reviews_all(
    sanjose311,
    sleep_milliseconds=10, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
seeclickfix_results = reviews_all(
    seeclickfix,
    sleep_milliseconds=10, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
findit_df = pd.DataFrame([{"rating" : review['score'], "review": review['content']} for review in finditfixit_results])
sj311_df = pd.DataFrame([{"rating" : review['score'], "review": review['content']} for review in sj311_results])
seeclick_df = pd.DataFrame([{"rating" : review['score'], "review": review['content']} for review in seeclickfix_results])
```

```python
findit_df.to_csv(f"../../initial_needfinding/data/reviews/google/FindItFixit.csv", index=False)
sj311_df.to_csv(f"../../initial_needfinding/data/reviews/google/SanJose311.csv", index=False)
seeclick_df.to_csv(f"../../initial_needfinding/data/reviews/google/SeeClickFix.csv", index=False)
```
